```r
likes <- c(40, 50, 60, 10, 30)
types <- c("Comedy", "Action", "Romance", "Drama", "SiFi")
```

```r
pie(likes, labels=types, main = "Favorite type of Movie", col=rainbow(length(likes)))
```

```r
percentages <- round((likes/sum(likes))*100, 1)
```

```r
pie(likes, labels = paste(percentages, "%"), col=rainbow(length(likes)), main="Movie Type Distribution")
```

```r
barplot(likes, names.arg=types, col="red", border="yellow",
        xlab="Movie Type", ylab="No of Likes", main="Favourite type of Movie")
```

```r
heart <-read.csv("heart.csv")
```

```r
games<-read.csv("games.csv")
```

```r
boxplot(heart$MaxHR ~ heart$Sex, 
        col = c("purple", "yellow"), 
        xlab = "Sex", 
        ylab = "Maximum Heart Rate", 
        main = "Heart Rates")
```

```r
print(boxplot.stats(heart$MaxHR[heart$Sex == "M"])$out)
```

```r
print(boxplot.stats(heart$MaxHR[heart$Sex == "F"])$out)
```

```r
x <- skewness(heart$MaxHR[heart$Sex == "M"])
if (x < 0){
    print("Left Skewness")
} else {
    
    print("Right Skewness")
}
```

```r
skewness(heart$MaxHR[heart$Sex == "F"])
if (x < 0){
    print("Left Skewness")
} else {
    
    print("Right Skewness")
}
```

```r
boxplot(games$turns ~ games$winner, 
        col = c("green", "blue", "red"), 
        xlab = "Winner of the game", 
        ylab = "No of turns", 
        main = "Chess game summary")
```

```r
boxplot.stats(games$turns[games$winner == "white"])$out
boxplot.stats(games$turns[games$winner == "black"])$out
boxplot.stats(games$turns[games$winner == "draw"])$out
```

```r
x = skewness(games$turns[games$winner == "white"])
if (x < 0){
    print("Left Skewness")
} else {
    
    print("Right Skewness")
}
x = skewness(games$turns[games$winner == "black"])
if (x < 0){
    print("Left Skewness")
} else {
    
    print("Right Skewness")
}
x = skewness(games$turns[games$winner == "draw"])
if (x < 0){
    print("Left Skewness")
} else {
    
    print("Right Skewness")
}
```
